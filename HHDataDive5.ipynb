```python
fig = plt.figure(figsize= (10,10))
sns.heatmap(df3.corr(), annot=True)
plt.show()
```

```python
# HDI vs GDPpc

plt.style.use('dark_background')

fig, ax = plt.subplots()
HDI_GDP = df3.plot(kind='scatter', x='GDPpc', y='HDI', c='expyrsskl', cmap='viridis', ax=ax)
# c.legend(["Expected Years of School"])
HDI_GDP.set_xlabel("GDP per capita")
HDI_GDP.set_ylabel("HDI")

fig, ax = plt.subplots()
HDI_GDP2 = df3.plot(kind='scatter', x='GDPpc', y='HDI', c='mnyrsskl', cmap='viridis', ax=ax)
# c.legend(["Expected Years of School"])
HDI_GDP2.set_xlabel("GDP per capita")
HDI_GDP2.set_ylabel("HDI")
```

```python
# HDI vs lex

plt.style.use('dark_background')
fig, ax = plt.subplots()
HDI_Lex = df3.plot.scatter(x = "lex", y= "HDI", c="FreedomToMakeLifeChoices", colormap = 'plasma', ax=ax)
HDI_Lex.set_xlabel("Life Expectancy")
HDI_Lex.set_ylabel("HDI")
```

```python
# HDI vs social support
plt.style.use('dark_background')

fig, ax = plt.subplots()
HDI_SS = df3.plot(kind='scatter', x='SocialSupport', y='HDI', c='WHRhappiness', cmap='Purples', ax=ax)
HDI_SS.set_xlabel("Social Support")
HDI_SS.set_ylabel("HDI")
```
